# 💼 Employee Salary Prediction — Questions



## 🟢 Basic Questions

**1. What is this project about?**
It's a supervised machine learning project that predicts an employee's salary based on inputs like experience, education, job role, location, and company size.

**2. What type of ML problem is this — regression or classification?**
Regression, because the target (salary) is a continuous numeric value, not a category.

**3. What's the difference between regression and classification?**
Regression predicts a continuous number (e.g., salary), while classification predicts a discrete label (e.g., "high/medium/low" salary band).

**4. What is supervised learning?**
Learning from labeled data — the model is trained on input-output pairs (features → known salary) so it can predict outputs for new, unseen inputs.

**5. What features did you use to predict salary?**
Experience, Education level, Job Role, Location, and Company Size.

**6. Why do we split the data into training and test sets?**
To evaluate how well the model performs on data it hasn't seen, which gives an honest measure of real-world performance.

**7. What tools/libraries did you use?**
Python, pandas and numpy for data handling, matplotlib/seaborn for visualization, and scikit-learn for modeling and evaluation.

**8. What is overfitting and underfitting, in simple terms?**
Overfitting is when a model memorizes the training data and fails on new data; underfitting is when a model is too simple to capture the actual pattern.

## 🟡 Intermediate Questions

**9. How did you handle categorical features like Job Role and Education?**
Education has a natural order, so it was ordinally encoded (e.g., High School < Bachelors < Masters < PhD). Job Role, Location, and Company Size have no inherent order, so they were one-hot encoded.

**10. Why is feature scaling important, and which models need it?**
Scaling brings all numeric features to a comparable range. Distance/gradient-based models like Linear Regression, KNN, and SVR need it; tree-based models like Random Forest don't, since they split on thresholds rather than distances.

**11. Which evaluation metrics did you use, and why those specifically?**
MAE (average absolute error, easy to interpret in salary units), RMSE (penalizes large errors more), and R² (how much variance in salary the model explains). Using multiple metrics avoids a misleading single-number conclusion.

**12. Why compare more than one algorithm?**
Different models make different assumptions — Linear Regression assumes linear relationships, while Random Forest can capture non-linear patterns. Comparing them shows whether the added complexity is actually worth it.

**13. What is cross-validation, and why use it over a single train/test split?**
Cross-validation splits the data into multiple folds and trains/evaluates the model several times on different subsets, giving a more reliable estimate of performance than a single split.

**14. What is multicollinearity, and how would you detect it here?**
When two or more features are highly correlated with each other (e.g., Experience and Age), which can distort coefficient interpretation in linear models. It can be checked using a correlation matrix or VIF (Variance Inflation Factor).

**15. What is the bias-variance tradeoff?**
Bias is error from overly simplistic assumptions (underfitting); variance is error from sensitivity to training data noise (overfitting). A good model balances both for the best generalization.

## 🔴 Advanced Questions

**16. How would you handle outliers in salary data (e.g., a CEO's salary skewing results)?**
Detect them using IQR or z-scores, then either cap/transform them (e.g., log-transform salary), remove them if they're data errors, or use models robust to outliers like tree-based ensembles.

**17. What is regularization, and why might you use Ridge or Lasso instead of plain Linear Regression?**
Regularization adds a penalty for large coefficients to reduce overfitting. Ridge (L2) shrinks coefficients smoothly; Lasso (L1) can shrink some coefficients to exactly zero, effectively performing feature selection.

**18. How does Random Forest actually work, and why does it tend to outperform a single Decision Tree?**
It builds many decision trees on random subsets of data and features, then averages their predictions. This reduces variance and overfitting compared to a single deep tree, which tends to memorize the training data.

**19. What's the difference between GridSearchCV and RandomizedSearchCV for hyperparameter tuning?**
GridSearchCV exhaustively tries every combination in a defined grid (slower but thorough); RandomizedSearchCV samples a fixed number of random combinations (faster, useful when the search space is large).

**20. How would you handle a categorical feature with very high cardinality, like hundreds of unique job titles?**
One-hot encoding would create too many sparse columns. Better approaches include target/mean encoding, frequency encoding, grouping rare categories into "Other," or using embeddings if using a neural network.

**21. How do you interpret feature importance in a Random Forest model, and what are its limitations?**
Feature importance shows how much each feature reduces prediction error on average across trees. Limitation: it can be biased toward high-cardinality features, so techniques like permutation importance or SHAP values give a more reliable picture.

**22. What is data leakage, and how could it happen in this pipeline if you're not careful?**
Data leakage is when information from outside the training set (often from the test set) influences model training, inflating performance unrealistically. Example: fitting the `StandardScaler` on the entire dataset before splitting, instead of fitting it only on the training set.

**23. How would you deploy this model into production, and what would you monitor afterward?**
Wrap the trained model in an API (e.g., Flask/FastAPI) or app (Streamlit), serve predictions, and monitor for data drift (if incoming employee data starts looking different from training data) and model drift (performance degrading over time), retraining periodically.

**24. What's the difference between L1 and L2 regularization?**
L1 (Lasso) adds the absolute value of coefficients as a penalty and can shrink some to zero (sparse model, built-in feature selection). L2 (Ridge) adds the squared value of coefficients and shrinks them smoothly toward zero without eliminating any.

**25. If your model performs very well on training data but poorly on test data, what would you do?**
This signals overfitting. Steps include using cross-validation, simplifying the model, adding regularization, gathering more data, or reducing the number of features to improve generalization.

**26. How would you explain a single prediction from your model to a non-technical stakeholder?**
Using interpretability tools like SHAP or LIME, which break down a single prediction into the contribution of each feature (e.g., "+$8,000 for Masters degree, +$12,000 for Tier-1 city location"), making the "black box" prediction explainable in plain terms.
